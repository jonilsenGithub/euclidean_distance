# Euclidean Distance

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
original_df = pd.read_csv('keystroke.csv')
original_df
```

```python
original_df['subject'] = org_df['subject'].apply(lambda x: int(x[2:]))
```

```python
original_df.columns
```

## Splitting into training set and test set

```python
training_set = original_df[org_df['sessionIndex']!=8 ]
test_set = original_df[org_df['sessionIndex']==8]
```

## Splitting further into hold times and up-down times

```python
training_hold_df = training_set[['subject','H.period','H.t','H.i','H.e','H.five','H.Shift.r','H.o','H.a','H.n', 'H.l']]
training_hold_df = training_hold_df.reset_index().drop('index',axis=1)

training_ud_df = training_set[['subject','UD.period.t','UD.t.i','UD.i.e','UD.e.five','UD.five.Shift.r','UD.Shift.r.o','UD.o.a','UD.a.n','UD.n.l','UD.l.Return']]
training_ud_df = training_ud_df.reset_index().drop('index',axis=1)

test_hold_df = test_set[['subject','H.period','H.t','H.i','H.e','H.five','H.Shift.r','H.o','H.a','H.n', 'H.l']]
test_hold_df = test_hold_df.reset_index().drop('index',axis=1)

test_ud_df = test_set[['subject','UD.period.t','UD.t.i','UD.i.e','UD.e.five','UD.five.Shift.r','UD.Shift.r.o','UD.o.a','UD.a.n','UD.n.l','UD.l.Return']]
test_ud_df = test_ud_df.reset_index().drop('index',axis=1)
```

## Creating Threshold Profiles

```python
training_hold_df
```

```python
from sklearn.metrics.pairwise import euclidean_distances

training_euc = []

for row in range(0,17850):
    if training_hold_df.iloc[row]['subject'] == training_ud_df.iloc[row]['subject']:
        training_euc.append(euclidean_distances(np.array(training_hold_df.drop('subject',axis=1).iloc[row]).reshape(1,-1),
                        np.array(training_ud_df.drop('subject',axis=1).iloc[row]).reshape(1,-1)))
```

```python
training_euc = np.array(training_euc).reshape(-1,1)
training_hold_df['euc'] = training_euc
```

```python
euc_profiles = training_hold_df[['subject','euc']]
euc_profiles = euc_profiles.groupby('subject').mean()
```

```python
euc_profiles['min_tresh'] = euc_profiles['euc']-0.04
euc_profiles['max_tresh'] = euc_profiles['euc']+0.04
```

```python
euc_profiles = euc_profiles.reset_index()
```

```python
euc_profiles
```

```python
test_hold_df = test_hold_df.reset_index().drop('index',axis=1)
test_ud_df = test_ud_df.reset_index().drop('index',axis=1)
```

## Classifying test set

```python
import random
from sklearn.metrics.pairwise import euclidean_distances

count = 0

for i in range(0,100):
    
    rand = random.randint(0,2549) #random int between 0 and 2549 as test set has 2550 rows
    
    #choosing random sample among hold samples
    hold = np.array(test_hold_df.iloc[rand]).reshape(1,-1)
    hold_profile = np.delete(hold,0,axis=1) #extracting features
    hold_subject = hold[0][0] #extracting sample user
    
    #choosing random sample among up-down samples
    ud = np.array(test_ud_df.iloc[rand]).reshape(1,-1)
    ud_profile = np.delete(ud,0,axis=1) #extracting features
    ud_subject = ud[0][0] #extracting sample user
    
    #retrieving random user's actual threshold profile
    actual_profile = np.array(euc_profiles[euc_profiles['subject']==hold_subject].drop('subject',axis=1))
    
    #checking euclidean distance between hold sample and up-down sample to see if it is in the profile threshold
    if euclidean_distances(hold_profile,ud_profile)>actual_profile[0][1] and euclidean_distances(hold_profile,ud_profile)<actual_profile[0][2]:

        count = count + 1
        
print('Accuracyscore is {}'.format(count/100))
```
